Requirement

1.	Group the table by Educational Attainment and sum the Population Count in each category.

3.	Using pivot , get a contingency table (a table of counts) of adult Californians cross-classified by Educational Attainment and Personal Income.
(see the power of pivot over other cross-classification methods. Each column of counts is a distribution of personal income at a specific level of educational attainment.Converting the counts to percents allows us to compare the four distributions).

            
(Note: Over 35% of those with Bachelor's deg or higher had incomes of  75K and over, Vs fewer than 10% of the people in the other education categories had that level of income.)

4.	Draw a bar chart to compare the personal income distributions of adult Californians who have no high diploma with those who have completed a Bachelor's degree or higher.  (The difference in the distributions would be striking. There is a clear positive association between educational attainment and personal income).


```python
#3
import pandas as pd
q3=pd.read_csv('https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv')
q3
```

```python
q3=q3[q3.Year>='1/1/14']
q3=q3[q3.Age>'127']
q3=q3.rename(columns={'Population Count':'Pcount'})
```

```python
q3.groupby('Educational Attainment').Pcount.sum()
```

```python
count=q3['Pcount'].sum()
q3.groupby('Educational Attainment').Pcount.sum()*100/count
```

```python
table = pd.pivot_table(q3,index = ['Educational Attainment','Personal Income'])
table.sort_values(by = 'Educational Attainment', ascending = False, inplace = True)
table.drop(['College, less than 4-yr degree', 'High school or equivalent'], inplace = True)
table.reset_index('Personal Income')
```

```python
table = table.unstack()
q3 = pd.DataFrame(table)                                                                                     
q3 = q3.T
q3.sort_values(by = 'Personal Income',ascending = False, inplace = True)
q3.plot.barh(color = {'No high school diploma': 'yellow', "Bachelor's degree or higher": 'blue'})
q3
```
